```python
train_ham_data=pd.read_csv('H3DS5230Data/flagTrainHam.csv')
train_ham_data = train_ham_data.drop(['Unnamed: 0'], axis=1)
train_ham_data.head()
```

```python
#Load train spam data
train_spam_data = pd.read_csv('H3DS5230Data/flagTrainSPAM.csv')
train_spam_data = train_spam_data.drop(['Unnamed: 0'], axis=1)
train_spam_data.head()
```

```python
#Load test ham data
test_ham_data = pd.read_csv('H3DS5230Data/flagTestHam.csv')
test_ham_data = test_ham_data.drop(['Unnamed: 0'], axis=1)
test_ham_data.head()
```

```python
#Load test spam data
test_spam_data = pd.read_csv('H3DS5230Data/flagTestSPAM.csv')
test_spam_data = test_spam_data.drop(['Unnamed: 0'], axis=1)
test_spam_data.head()
```

# 5c

```python
train_ham_data_sum = train_ham_data.sum(skipna = True)
train_ham_MLE = train_ham_data_sum/len(train_ham_data)
train_ham_MLE
```

```python
train_spam_sum = train_spam_data.sum( skipna = True)
train_spam_MLE = train_ham_data_sum/len(train_spam_data)
train_spam_MLE
```

```python
train_ham_data_sum_temp = train_ham_data.sum(skipna = True)
map_of_temp_ham = []
for i in range(len(train_ham_data_sum_temp)):
    map_of_new_ham = (train_ham_data_sum_temp[i]+2-1)/(len(train_ham_data)+2+2-2)
    map_of_temp_ham.append(map_of_new_ham)
```

```python
map_of_temp_ham
```

```python
train_spam_data_sum_temp = train_spam_data.sum(axis = 0, skipna = True)
map_temp_data_spam = []
for i in range(len(train_spam_data_sum_temp)):
    map_new_spam = (train_spam_data_sum_temp[i]+2-1)/(len(train_spam_data)+2+2-2)
    map_temp_data_spam.append(map_new_spam)
```

```python
map_temp_data_spam
```

```python
y_ham=train_ham_MLE.tolist()
y_spam = train_spam_MLE.tolist()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
# Left hand side plot
plt.subplot(1,2,1)
plt.plot(range(train_ham_data.shape[1]),y_ham,'ro',label='MLE')
plt.plot(range(train_ham_data.shape[1]),map_of_temp_ham,'go',label='MAP')
plt.title('Scatter Plot Ham')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
# Right hand side plot
plt.subplot(1,2,2)
plt.plot(range(train_spam_data.shape[1]),y_spam,'ro',label='MLE')
plt.plot(range(train_ham_data.shape[1]),map_temp_data_spam,'go',label='MAP')
plt.title('Scatter Plot Spam')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

# 5d

```python

for i in range(len(train_spam_data_sum_temp)):
#Hits
    log_hit_prob_ratio = np.log(np.array(map_temp_data_spam)/np.array(map_of_temp_ham))
#Miss
    log_miss_prob_ratio = np.log((1- np.array(map_temp_data_spam))/(1-np.array(map_of_temp_ham)))
    log_prob_ratio_test = test_spam_data.sum(axis = 1, skipna = True)
```

```python
log_hit_prob_ratio
```

```python
#Plotting the histogram
plt.figure(figsize=(10,4))
# Left hand side plot
plt.subplot(1,2,1)
n,bins,patches = plt.hist(x=log_hit_prob_ratio,bins=100,color='green')
plt.title('Histogram with Spam')
plt.xlabel('X')
plt.ylabel('Y')
# Right hand side plot
plt.subplot(1,2,2)
n,bins,patches = plt.hist(x=log_miss_prob_ratio,bins=100,color='red')
plt.title('Histogram with Ham')
plt.xlabel('X')
plt.ylabel('Y')
```
